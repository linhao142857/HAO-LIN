```python
#present_value function gives the discounted value of all remaining coupons and principal
def present_value (rfrs, spr, c, m, n):
    
    #rsf is risk free rate series(from 0.25 year to 30.00 years)
    #spr is spread
    #c is coupon rate 
    #m is the total number of coupons
    #n is number of coupon payments per year         
            
    denominator = np.zeros(m)
    for i in range(0, m):
        denominator[i] = denominator[i] + (1 + spr + (rfrs[int(4*(i+1)/n) -1])/n)**(i+1)
        #denominator is a numpy arrary that contains all denominators
    
    p0 = sum(c/denominator) + 1/denominator[-1]
    return p0*100    
```

```python
def bisection(f,a,b,N):
    
    '''
    Parameters
    ----------
    f : function
        The function for which we are trying to approximate a solution f(x)=0.
        In this case, it is present value - p0 = 0
    a,b : numbers
        The interval in which to search for a solution. 
    N : (positive) integer
        The number of iterations to implement. 
    -------
    '''
    #check boundary value
    if abs(f(a)) <= 0.007: #I can't go smaller and higher than 0.007 might not give the precise result
        return a
    if abs(f(b)) <= 0.007:
        return b    
    
    while f(b) >= 0.007:
        b = b - 1
        if abs(f(b)) <= 0.007:
            return b
        if b<2:
            v = np.zeros(b-a)
            for i in range(0,b-a):
                v[i] = v[i] + abs(f(i+a))
            return (np.argmin(v)+ a)
         
    a_n = a
    b_n = b
                      
    for n in range(1,N+1):
        
        m_n = round((a_n + b_n)/2)
            
        if abs(f(m_n)) <= 0.007:
            return m_n
        if abs(f(m_n - 1)) <= 0.007:
            return (m_n - 1)
        if abs(f(m_n + 1)) <= 0.007:
            return (m_n + 1)
        
        #this is from tuturial 2, but this is not the case for some bonds(not sure)
        if f(m_n) < 0:
            a_n = a_n
            b_n = m_n
        
        elif f(m_n) >= 0:
            a_n = m_n
            b_n = b_n
        
        else:
            if round((a_n+b_n)/2) == (a_n+ b_n)/2:
                return (round((a_n+b_n)/2))
            else:
                if abs(f(int((a_n + b_n)/2 + 0.5))) >= abs(f(int((a_n + b_n)/2 - 0.5))):
                    return (int((a_n + b_n)/2 - 0.5))
                elif abs(f(int((a_n + b_n)/2 + 0.5))) <= abs(f(int((a_n + b_n)/2 - 0.5))):
                        return (int((a_n + b_n)/2 + 0.5))
                else:
                    return None
```

```python
#function g applies bisection function above on f, with 3 cases.
def g(row):
    rfrsr = row[" 025YR" : " 3000YR"]
    sprr  = row["Spread"]
    nr  = row["Number"]
    cr =    row["CouponRate"]
    f = lambda x: present_value(rfrsr, sprr, cr, x , nr) - row["PriceAtCoupon0(%)"]
    if nr == 1:
        return(bisection(f,1,30,6))
    if nr == 2: 
        return(bisection(f,1,60,7))
    if nr == 4:
        return(bisection(f,1,120,8))
```

```python
import datetime
from datetime import date
from dateutil.relativedelta import relativedelta
```

```python
#this function is to adjust the last coupon price up to real Aug 25 price, which is the dirty price
def adjustfactor(row):
    date = row['Date']
    current = row['CurrentCouponNumbers']
    today = datetime.datetime(2021, 8, 25)
    n = row['Number']
    spr = row['Spread']
    if n == 1:
        last_coupon_date = date + relativedelta(months = +current*12)
        delta = (today -last_coupon_date).days
        adjust_factor = (1 + spr + rfrs_aug_25[3])**(delta/365)
        
    elif n == 2:
        last_coupon_date = date + relativedelta(months = +current*6)
        delta = (today -last_coupon_date).days
        adjust_factor = (1 + spr + rfrs_aug_25[1]/2)**(delta/183)
        
    elif n == 4:
        last_coupon_date = date + relativedelta(months = +current*3)
        delta = (today -last_coupon_date).days
        adjust_factor = (1 + spr + rfrs_aug_25[0]/4)**(delta/91)
    return(adjust_factor)
```

```python
dfb['Adjustfactor'] = dfb.apply(adjustfactor, axis = 1)
dfb.iloc[: , -1]
```

```python
dfb['Aug25DirtyPrice'] = dfb['Aug25PriceAtLastCoupon'] * dfb['Adjustfactor']
dfb.iloc[: , -1]
```

```python
#Actualprice is adjusted with FaceValue for each bond, will be used in 1(c)
dfb['Aug25DirtyPriceWithFV'] = dfb['Aug25DirtyPrice'] * dfb['FaceValue']/100
dfb.iloc[: , -1]
```

```python
#5 percentile annualized spread is 0.0112020499013506
#20 percentile annualized spread is 0.0337962821890598
#40 percentile annualized spread is 0.058478470289440414
#60 percentile annualized spread is 0.0843598271209384
#75 percentile annualized spread is 0.1175060520461992
#90 percentile annualized spread is 0.17286558519757517
def PD(AnnualizedSpread):
    if AnnualizedSpread < 0.0112020499013506:   #rating AAA
        PD = 0.001 
    elif AnnualizedSpread < 0.0337962821890598:    #rating AA
        PD = 0.01
    elif AnnualizedSpread < 0.058478470289440414:    #rating A
        PD = 0.12
    elif AnnualizedSpread < 0.0843598271209384:   #rating BBB
        PD = 0.36
    elif AnnualizedSpread < 0.1175060520461992:  #rating BB
        PD = 0.52
    elif AnnualizedSpread < 0.17286558519757517:   #rating B
        PD = 0.75
    #if SpreadDecile >= 0.9:  #rating CCC
    else:
        PD = 0.85
    return PD
```

```python
#assign PD to each bond based on their spreaddecile
dfb['PD'] = dfb.apply(lambda x: PD(x['AnnualizedSpread']),axis=1)
dfb.iloc[: , -1]
```

```python
#calculate b
from scipy.stats import norm
def find_b(PD):
  b = (0.11852-0.05487*np.log(PD))**2
  return b
```

```python
dfb['b'] = dfb.apply(lambda x: find_b(x['PD']), axis = 1)
dfb.iloc[: , -1]
```

```python
#I think I remember we should treat all bonds as subordinated, so LGD = 0.75
dfb['LGD'] = dfb.apply(lambda x: 0.75, axis = 1)
dfb.iloc[: , -1]
```

```python
def capital_requirement_corporate(PD,LGD,b,M):
  # First part of the equation, lower correlation
  R =  0.12 * ( (1 - np.exp(-50 * PD)) / (1 - np.exp(-50)) )
  # Second part of the equation, higher correlation 
  R += 0.24 * (1 - ( (1 - np.exp(-50 * PD)) / (1 - np.exp(-50)) ) )
  # Now we can calculate the capital
  K = norm.cdf(np.sqrt( (1 - R) ** (-1) ) * norm.ppf(PD) + 
               np.sqrt( R / (1 - R) ) * norm.ppf(0.999) ) - PD
  K *= LGD*((1+(M-2.5)*b)/(1-1.5*b))
  return K
```

```python
#this should take about 1 min to run
dfb['Capi_req_Percent'] = dfb.apply(lambda x: capital_requirement_corporate(x.PD, x.LGD, x.b, x.TotalCoupons),axis = 1)
dfb.iloc[: , -1]
```

```python
dfb['Capi_req_amount'] = dfb.apply(lambda x: x.FaceValue * x.Capi_req_Percent,axis = 1)
dfb.iloc[: , -1]
```

```python
Total_Capital_requirement = np.sum(dfb.Capi_req_amount)
Total_Capital_requirement#.round(3)
```
